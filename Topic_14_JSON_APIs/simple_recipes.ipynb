```python
# we will be accessing https://dummyjson.com/recipes
# and getting the json data from the website
import requests
url = "https://dummyjson.com/recipes"
response = requests.get(url)
if response.status_code == 200:
    data = response.json()
    print(data)
```

```python
# so if this is a dictionary let's see the keys
print(data.keys())
```

```python
# so we will want data from recipes
recipes = data['recipes']
# how many recipes are there
print(len(recipes))
# so recipes is a list of dictionaries and you can do the rest of assignment
```

```python
# let's iterate over all recipes and save potatoe recipes in a new list
pototoe_recipes = []
needle = 'potato'
for recipe in recipes:
    # now I will go through ingredients and see if potatoe is in there
    for ingredient in recipe['ingredients']:
        if needle in ingredient.lower():
            pototoe_recipes.append(recipe)
            break # unlikely that we have two potatoe ingredients in one recipe
print(f"Found {len(pototoe_recipes)} recipes with potatoe")
```

```python
# let's print all 3 recipes
for recipe in pototoe_recipes:
    print(recipe['name'])
    print(recipe['ingredients'])
    print(recipe['instructions'])
    # we could find all keys and print them but this is enough for now
```

```python
from collections import Counter

# Get all ingredients from the recipes
all_ingredients = []
for recipe in recipes:
    all_ingredients.extend(recipe['ingredients'])

# Count the frequency of each ingredient
ingredient_counts = Counter(all_ingredients)

# Get the top 5 most common ingredients
top_ingredients = ingredient_counts.most_common(5)

# Print the top 5 ingredients
print("Top 5 most used ingredients:")
for ingredient, count in top_ingredients:
    print(f"{ingredient}: {count} times")
```

```python
# 1.e uzzīmējiet histogrammu ar TOP 10 ingredients skaitu 
import matplotlib.pyplot as plt

# Get the top 10 ingredients and their counts
top_ingredients = ingredient_counts.most_common(10)
ingredients = [ingredient[0] for ingredient in top_ingredients]
counts = [ingredient[1] for ingredient in top_ingredients]

# Plot the histogram
plt.bar(ingredients, counts)
plt.xlabel('Ingredients')
plt.ylabel('Count')
plt.title('Top 10 Ingredients Count')
plt.xticks(rotation=45)
plt.show()
```
